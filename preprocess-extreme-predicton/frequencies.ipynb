```python
def load_labels(path):
    print('Loading "%s" ...' % path)
    with open(path, 'rb') as f:
        num_lines, num_categories = np.fromfile(f, dtype=np.int32, count=2)
        num_bits = np.ceil(np.log2(num_categories))
        print('%d data points over %d (<= 2^%d = %d) categories'
              % (num_lines, num_categories, num_bits, 2**num_bits))
        labels = np.fromfile(f, dtype=np.int32, count=num_lines)
        # Make sure that we've reached EOF
        assert len(f.read(1)) == 0

    counts = np.zeros((num_categories,), np.int32)
    for i in labels:
        counts[i] += 1
    
    plt.loglog(counts)
```
